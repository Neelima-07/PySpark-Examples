```
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('SparkByExamples.com').getOrCreate()

data = ["Project Gutenberg’s",
        "Alice’s Adventures in Wonderland",
        "Project Gutenberg’s",
        "Adventures in Wonderland",
        "Project Gutenberg’s"]
rdd=spark.sparkContext.parallelize(data)

for element in rdd.collect():
    print(element)

#Flatmap    
rdd2=rdd.flatMap(lambda x: x.split(" "))
for element in rdd2.collect():
    print(element)
```

```
#Importing Libraries: The code imports the necessary library, pyspark.sql and SparkSession.

#Creating SparkSession: A Spark session is created using SparkSession.builder.appName('SparkByExamples.com').getOrCreate().

#Creating RDD: An RDD named rdd is created from a list of strings data using spark.sparkContext.parallelize(data).

#Iterating and Printing RDD Elements: The RDD rdd is iterated using a for loop, and each element is printed using print(element).

#Applying FlatMap Transformation: A new RDD named rdd2 is created by applying the flatMap transformation on rdd. The flatMap transformation splits each string element of rdd into multiple words by splitting on whitespace.

#Iterating and Printing FlatMapped RDD Elements: The RDD rdd2 is iterated using a for loop, and each element is printed using print(element).

#The code demonstrates the difference between rdd and rdd2. The original RDD rdd contains individual strings, while the flatMap transformation applied on rdd using x.split(" ") splits each string into multiple words, resulting in a flattened RDD rdd2 that contains individual words.

#By printing the elements of rdd and rdd2, you can observe the difference in their structures. The flatMap transformation is useful when you want to split elements or create multiple output elements from each input element in an RDD.
```
