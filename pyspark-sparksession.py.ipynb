```
import pyspark
from pyspark.sql import SparkSession
spark = SparkSession.builder.master("local[1]") \
                    .appName('SparkByExamples.com') \
                    .getOrCreate()

print("First SparkContext:");
print("APP Name :"+spark.sparkContext.appName);
print("Master :"+spark.sparkContext.master);

sparkSession2 = SparkSession.builder \
      .master("local[1]") \
      .appName("SparkByExample-test") \
      .getOrCreate();

print("Second SparkContext:")
print("APP Name :"+sparkSession2.sparkContext.appName);
print("Master :"+sparkSession2.sparkContext.master);


sparkSession3 = SparkSession.builder.getOrCreate()

print("Third SparkContext:")
print("APP Name :"+sparkSession3.sparkContext.appName);
print("Master :"+sparkSession3.sparkContext.master);
```

```
#The code imports the necessary modules: pyspark and SparkSession from pyspark.sql.

#The first SparkSession is created using the SparkSession.builder API. The master parameter is set to "local[1]", which runs Spark locally with one worker thread. The appName parameter sets the name of the Spark application.

#The properties of the first SparkSession are accessed using the spark.sparkContext object. The appName property is accessed with spark.sparkContext.appName and the master property is accessed with spark.sparkContext.master. The values are then printed to the console.

#The second SparkSession is created using the same approach as the first one, with a different application name.

#The properties of the second SparkSession are accessed and printed to the console in the same way as in the previous step.

#The code attempts to create a new SparkSession named sparkSession3 using the newSession method. However, the newSession method is missing the parentheses () and will result in a syntax error.

#The properties of the non-existent sparkSession3 are accessed and printed to the console, which will result in an AttributeError since sparkSession3 was not successfully created.
```
