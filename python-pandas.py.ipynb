```
import pandas as pd    
data = [["James","","Smith",30,"M",60000], 
        ["Michael","Rose","",50,"M",70000], 
        ["Robert","","Williams",42,"",400000], 
        ["Maria","Anne","Jones",38,"F",500000], 
        ["Jen","Mary","Brown",45,None,0]] 
columns = ['First Name', 'Middle Name','Last Name','Age','Gender','Salary']

# Create the pandas DataFrame 
pandasDF = pd.DataFrame(data=data, columns=columns) 
  
# print dataframe. 
print(pandasDF)

#Outputs below data on console

pdCount=pandasDF.count()
print(pdCount)

print(pandasDF.max())
print(pandasDF.mean())
```

```
#The code begins by importing the Pandas library, which is a popular library for data manipulation and analysis in Python.

#The data variable is a list of lists that represents the data to be loaded into the DataFrame. Each inner list corresponds to a row of data, and the elements within the inner list represent the values of the columns in that row.

#The columns variable is a list that specifies the names of the columns in the DataFrame.

#The pd.DataFrame() constructor is used to create a DataFrame named pandasDF by passing the data and columns. This creates a tabular structure where the data is organized into rows and columns.

#The print(pandasDF) statement is used to display the contents of the DataFrame, which shows the data in a tabular form.

#The count() method is called on the DataFrame pandasDF to count the number of non-null values in each column. The result is stored in the pdCount variable.

#The print(pdCount) statement is used to display the counts of non-null values for each column, providing insights into the completeness of the data.

#The max() method is called on the DataFrame to find the maximum value in each column. This is useful for identifying the largest value in numerical columns.

#The print(pandasDF.max()) statement is used to display the maximum values for each column.

#The mean() method is called on the DataFrame to calculate the mean (average) value in each column. This is applicable to numerical columns.

#The print(pandasDF.mean()) statement is used to display the average values for each column.

#Overall, the code showcases the basic usage of Pandas DataFrames, including creating a DataFrame, printing its contents, counting non-null values, and calculating statistics such as the maximum and mean values.
```
